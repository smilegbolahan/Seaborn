```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('study2.csv')
data
```

```python
new_data = data.groupby(["Region"]).sum()
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x=data["Region"], y=data["Total population"], hue=data["Region"])
```

```python
plt.figure(figsize=(15,10))
sns.histplot(x=data["Region"], y=data["Total population"], hue=data["Region"])
```

```python
plt.figure(figsize=(15,10))
sns.stripplot(x=data["Region"], y=data["Total population"], hue=data["Region"])
```
