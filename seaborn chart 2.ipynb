```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("study2.csv")
```

```python
data
```

```python
data["Total population"].str.replace("," , "")
```

```python
data["Total population"] = data["Total population"].str.replace("," , "").astype(int)
```

```python
data
```

```python
new_data = data.groupby(["Region"]).sum()
```

```python
new_data
```

```python
new_data["Region"] = ["North Central", "North East", "North West", "South East", "South South", "South West"]
```

```python
new_data
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=new_data["Region"], y=sorted(new_data["Total population"]), data=new_data)
plt.text(0, new_data["Total population"][0], "29252408")
plt.text(1, new_data["Total population"][1], "26263866")
plt.text(2, new_data["Total population"][2], "50942307")
plt.text(3, new_data["Total population"][3], "21955414")
plt.text(4, new_data["Total population"][4], "28829288")
plt.text(5, new_data["Total population"][5], "37707260")
```

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid")
sns.scatterplot(x=new_data["Region"], y=sorted(new_data["Total Doses Administered (Moderna)"]), data=new_data, hue=new_data["Total Doses Administered (Moderna)"])
```

```python
plt.figure(figsize=(15,10))
sns.set_style("dark")
sns.histplot(x=new_data["Region"], y=sorted(new_data["Total Doses Administered (Moderna)"]), data=new_data, hue=new_data["Total Doses Administered (Moderna)"])
```

```python
plt.figure(figsize=(15,10))
sns.set_style("dark")
sns.stripplot(x=new_data["Region"], y=sorted(new_data["Total Doses Administered (Moderna)"]), data=new_data, hue=new_data["Total Doses Administered (Moderna)"])
```

```python
plt.figure(figsize=(15,10))
sns.set_style("dark")
sns.violinplot(x=new_data["Region"], y=sorted(new_data["Total Doses Administered (Moderna)"]), data=new_data, hue=new_data["Total Doses Administered (Moderna)"])
```

```python
plt.figure(figsize=(15,15))
sns.set_style("dark")
sns.displot(x=new_data["Region"], y=sorted(new_data["Total Doses Administered (Moderna)"]), data=new_data, hue=new_data["Total Doses Administered (Moderna)"])
```
